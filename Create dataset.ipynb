```python
import pandas as pd
import os
pd.options.display.max_rows = None #常に全ての列（カラム）を表示
pd.set_option('display.max_columns', None) #常に全ての列（カラム）を表示

datanme = '20190206 3D131収録リスト.xlsx'

datapath = os.path.join('data', datanme) 
dfdata = pd.read_excel(datapath)

dfdata.head()
```

```python
dfdata.iloc[0,3]
```

```python
# extract '3D131'
spld = dfdata.iloc[0,3].split(';')
print(spld)
```

```python
fterm = [s for s in spld if '3D131' in s]
print(fterm)
```

```python
idx = dfdata.iloc[0,0]
filename = idx + '_' + "_".join(fterm) + '.txt'
print(filename)
```

```python
text = dfdata.iloc[0,6] + ' ' + dfdata.iloc[0,7]
```
