```python
import pandas as pd
import os
pd.options.display.max_rows = None #常に全ての列（カラム）を表示
pd.set_option('display.max_columns', None) #常に全ての列（カラム）を表示

datanme = '20190206 3D131収録リスト.xlsx'
listname = '3D131_Fタームリスト.xlsx'

datapath = os.path.join('data', datanme) 
listpath = os.path.join('data', listname)
dfdata = pd.read_excel(datapath)
dflist = pd.read_excel(listpath)

dflist
```

```python
dfdata[dfdata['Ｆターム(最新)'].str.contains(dflist.iloc[2,1])]
```

```python
len(dfdata[dfdata['Ｆターム(最新)'].str.contains(dflist.at[0,'分類コード'])])
```

## frequency of f term

```python
dflist["freq"] = 0
for index, row in dflist.iterrows():
    dflist.at[index, 'freq'] = len(dfdata[dfdata['Ｆターム(最新)'].str.contains(dflist.at[index,'分類コード'])])
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt


colorlist = ["c", "m", "y"]

y = dflist.loc[1:,'freq']
x_label = dflist.loc[1:, '分類コード']
x = list(range(1, len(y)+ 1)) 

plt.barh(x, y, align="center", color=colorlist)
for xv in x:
    plt.text(50,xv,str(y[xv]), )

plt.yticks(x, x_label)
plt.show()
```

```python
# export graph
%matplotlib inline

x = list(range(1, len(y)+1)) 
plt.figure(figsize=(46.81, 33.11))
plt.rcParams["font.size"] = 5
plt.barh(x, y, align="center", color=colorlist)
for xv in x:
    plt.text(50,xv,str(y[xv]), ha='center', va='center')
plt.yticks(x, x_label)
plt.ylim(0, len(y)+1)
plt.subplots_adjust(left=0.1, right=0.95, bottom=0.1, top=0.95)
plt.savefig("sample.pdf",format = 'pdf', dpi=300)
```
