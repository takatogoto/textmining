# Create dataset

```python
import pandas as pd
import os
pd.options.display.max_rows = None #常に全ての列（カラム）を表示
pd.set_option('display.max_columns', None) #常に全ての列（カラム）を表示

datanme = '20190206 3D131収録リスト.xlsx'
setpath = os.path.join('data','set')

datapath = os.path.join('data', datanme) 
dfdata = pd.read_excel(datapath)

dfdata.head()
```

```python
for index,item in dfdata.iterrows():
    
    #if index > 1:
    #    continue
    
    # get list of the F-term
    spld = item['Ｆターム(最新)'].split(';')
    fterm = [s for s in spld if '3D131' in s]
    
    # 公報番号に'/'が入っているので'-'に置換
    idx = item['公報番号(全文リンク)'].replace('/', '-')
    
    listname = idx + '_fterm.txt'
    filename = idx + '_main.txt'
    
    listpath = os.path.join(setpath, listname)
    filepath = os.path.join(setpath, filename)
    
    
    ftermtxt = ", ".join(fterm)
    text = item['請求の範囲'] + ' ' + item['詳細な説明']
    
    
    with open(listpath, mode='w', encoding='utf-8') as f:
        f.write(ftermtxt)
        
    with open(filepath, mode='w', encoding='utf-8') as f2:
        f2.write(text)
```
